```python
import glob
import csv

import pandas as pd
```

```python
data_files = ["data/preference_data/mturk_5-31_batch.csv", "data/preference_data/mturk_6-1_batch.csv"]

mturk_df = pd.DataFrame()

columns = ["WorkerId"]

for i in range(1, 6):
    columns.append(f"Answer.pref{i}")
    columns.append(f"Input.model1_summary{i}")
    columns.append(f"Input.model2_summary{i}")
    columns.append(f"Input.id{i}")

for file_path in data_files:
    df = pd.read_csv(file_path, usecols=columns)
    mturk_df = pd.concat([mturk_df, df], ignore_index=True)
```

```python
mturk_df['WorkerId'] = pd.factorize(mturk_df['WorkerId'])[0] + 1
```
